```python
%%eigensheep --layer arn:aws:lambda:us-east-1:972882471061:layer:dna-tesseract:1 pytesseract requests --memory 3008
        
import pytesseract
import shutil
import requests
from PIL import Image

url = 'https://projectnaptha.com/img/tyger.jpg'
response = requests.get(url, stream=True)
with open('/tmp/img.png', 'wb') as out_file:
    shutil.copyfileobj(response.raw, out_file)
del response

print(pytesseract.image_to_string(Image.open('/tmp/img.png')))
```
